```python
train_titanic, validate_titanic, test_titanic = prepare.splitting_data(acquire.get_titanic_db(), 'survived')
train_titanic
```

```python
train_titanic = train_titanic.assign(is_male = np.where(train_titanic['sex'] == 'male', 1, 0))
train_titanic.head()
```

```python
train_titanic.info()
```

```python
train_titanic.embark_town.mode()
```

```python
train_titanic['embark_town'] = train_titanic.embark_town.fillna(train_titanic.embark_town.mode())
```

```python
train_titanic.embark_town.isnull().sum()
```

```python
train_titanic['embark_town'] = train_titanic['embark_town'].fillna(train_titanic.embark_town.mode())
```

```python
train_titanic['embark_town'] = train_titanic['embark_town'].fillna('Southampton')
```

```python
train_titanic.head()
```

```python
train_titanic = train_titanic.assign(is_southampton = np.where(train_titanic['embark_town'] == 'Southampton', 1, 0))
```

```python
train_titanic = train_titanic.assign(is_queenstown = np.where(train_titanic['embark_town'] == 'Queenstown', 1, 0))
```

```python
train_titanic.head()
```

```python
dummy_train_titanic = pd.get_dummies(train_titanic[['sex', 'embark_town']], drop_first=[True, True])
```

```python
dummy_train_titanic.head()
```

```python
dummy_train_titanic = pd.get_dummies(train_titanic[['sex', 'embark_town']], drop_first=[True, True]).astype(int)
dummy_train_titanic.head()
```

```python
train_titanic = pd.concat([train_titanic, dummy_train_titanic], axis=1)
train_titanic
```

```python
train_titanic = train_titanic.drop(columns=['sex', 'embark_town'])
```
